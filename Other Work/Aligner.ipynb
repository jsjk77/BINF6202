```python
residues = []
dssp = []
for i in lines[28:]:
    residues.append(i[13])
    dssp.append(i[16])
    print(i)
```

```python
dssp2 = []
for l in dssp:
    if l == " ":
        dssp2.append("C")
    else: 
        dssp2.append(l)
dssp2
```

```python
dssp1 = []
for m in dssp2:
    if m == "H" or m == "G" or m=="I":
        dssp1.append("H")
    elif m=="E" or m=="B":
        dssp1.append("E")
    elif m=="T" or m=="S" or m=="C":
        dssp1.append("C")
dssp1
```
